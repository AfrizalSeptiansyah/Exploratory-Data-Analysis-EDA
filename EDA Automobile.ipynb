## Cleaning Data
Pada proses ini kita akan melakukan pembersihan data yaitu : menghandle missing value/nilai kosong dan menghandle format data yang salah.

### Handle Missing Values 

```python
# mendeteksi missing values/nilai kosong pada data 
count_missing = data.isna().sum()
jml_missing = count_missing>0
print('Jumblah variabel yg missing :', jml_missing.sum())
count_missing
```

```python
# persentase dari data yg missing 
percentage = round((count_missing/len(data))*100,2)
percentage
```

Dapat kita lihat bahwa terdapat 10 variabel/fitur yang mengalami missing value. Namun, jika kita lihat tidak ada persentase missing value yg >=50% maka tidak ada variabel yang harus didrop/dihapus.

Adapun cara mengatasi missing value pada variabel-variabel tersebut adalah sebagai berikut :
* 'normalized-losses', 'wheel-base', 'bore', 'stroke', 'horsepower', 'peak-rpm', 'city-mpg' dan 'price' karena tipe datanya numerik maka boleh diisi dengan mean atau median. Disini saya memilih menggunakan median dari masing-masing variabel tersebut karena tidak terpengaruh oleh outlier. Untuk variabel price data yg missing akan kita drop karena price adalah target jadi tidak boleh ada yg kosong.
* 'num-of-doors' dan 'engine-location' karena tipe datanya kategorik diisi dengan modus dari masing-masing variabel tersebut. 

```python
# drop data yg kosong pd variabel price
data.dropna(subset=['price'], inplace=True, axis=0)
```

```python
# cek dimensi data setelah beberapa data didrop
data.shape
```

```python
# fill missing values pada variabel numerik 
features = ['normalized-losses','wheel-base','bore','stroke','horsepower', 
            'peak-rpm','city-mpg','price']

for i in features:
    if i in data.columns.values:
        data[i].fillna(data[i].astype('float64').median(axis=0), inplace=True)
```

```python
# fill missing values pada variabel kategorik 
features = ['num-of-doors','engine-location']

for i in features:
    if i in data.columns.values:
        data[i].fillna(data[i].value_counts().index[0], inplace=True)
```

```python
data.isna().sum()
```

Well, akhirnya data kita telah bersih dari missing values.

### Handle Incorrect Format Data

Sebelumnya kita telah mengetahui bahwa ada beberapa variabel yang format datanya tidak sesuai. Sekarang saya akan memperbaiki format data tersebut.

```python
features = ['normalized-losses','horsepower','peak-rpm']

for i in features:
    data[i] = data[i].astype('int64')
```

```python
features = ['bore','stroke','price']

for i in features:
    data[i] = data[i].astype('float64')
```

* Pada drive-wheels fwd mendominasi diikuti oleh rwd dengan perbedaan tidak begitu jauh.

### Bivariate Analysis

#### Korelasi 

```python
# corelation matrix 
plt.figure(figsize=(15,15))
sns.heatmap(data.corr(), annot=True, linewidths=0.5, cmap='RdYlGn')
plt.show()
```

* Variabel yang memiliki korelasi linear negatif kuat dengan price yaitu : highway-mpg dan city-mpg.
* Variabel yang memiliki korelasi linear positif kuat dengan price yaitu : horsepower, bore, engine-size, curb-weight, width, length dan wheel-base. 

```python
data.head()
```

#### Visualisasi variables yang memilki korelasi

```python
# korelasi negatif 
features = ['highway-mpg','city-mpg']

c = plt.figure(figsize=(10,8))
for i, j in enumerate(features):
    c.add_subplot(2,2,i+1)
    sns.regplot(x=data[j], y=data['price'])
    c.tight_layout()
```

* Dapat dilihat bahwa semakin rendah higway-mpg maka semakin tinggi price. Begitu pula yang terjadi dengan city-mpg. Itulah yang disebut sebagai korelasi linear negatif.

```python
# korelasi positif 
features = ['horsepower','bore','engine-size','curb-weight','width',
            'length','wheel-base']

d = plt.figure(figsize=(10,10))
for i, j in enumerate(features):
    d.add_subplot(3,3,i+1)
    sns.regplot(x=data[j], y=data['price'])
    d.tight_layout()
```

* Dapat dilihat bahwa semakin tinggi horsepower maka semakin tinggi price. Begitu pula yang terjadi dengan variabel-variabel lain jika dibandingkan dengan price. Itulah yang disebut sebagai korelasi linear positif.

#### Hubungan fuel-type, body-style, aspiration, drive-wheels dengan price

```python
features = ['fuel-type','body-style','aspiration','drive-wheels']

e = plt.figure(figsize=(10,8))
for i, j in enumerate(features):
    e.add_subplot(2,2,i+1)
    sns.barplot(x=data[j], y=data['price'], data=data)
    e.tight_layout()
```

* Untuk fuel-type diesel memiliki rata-rata harga lebih besar dibandingkan gas walaupun selisihnya tidak begitu jauh. Ini mungkin menjadi alasan mengapa lebih banyak yang menggunakan gas dibanding diesel.
* Untuk body-style convertible dan hardtop memiliki rata-rata harga lebih besar dibandingkan dengan yang lainnya walaupun diantara keduanya memiliki perbedaan rata-rata yang sangat kecil.
* Untuk aspiration turbo memiliki rata-rata harga lebih besar dibandingkan std.
* Untuk drive-wheels rwd memiliki rata-rata harga yang cukup jauh dibandingkan fwd dan 4wd.

```python
# boxplot
features = ['fuel-type','body-style','aspiration','drive-wheels']

f = plt.figure(figsize=(10,8))
for i, j in enumerate(features):
    f.add_subplot(2,2,i+1)
    sns.boxplot(x=data[j], y=data['price'], data=data)
    f.tight_layout()
```

* Hanya gas dan std yang memiliki outlier yang cukup banyak. Dan selebihnya memiliki outlier yang bisa dikatakan sedikit.

```python
# violinplot
features = ['fuel-type','body-style','aspiration','drive-wheels']

f = plt.figure(figsize=(10,8))
for i, j in enumerate(features):
    f.add_subplot(2,2,i+1)
    sns.violinplot(x=data[j], y=data['price'], data=data)
    f.tight_layout()
```

* Dengan menggunakan violinplot kita dapat melihat bentuk distribusi dari setiap kategori pada masing-masing variabel. 
* Titik putih ditengah merupakan median dibawahnya adalah Q1 dan diatasnya adalah Q3. Seperti lidi pada bagian atas merupakan batas atas dan pada bagian bawah merupakan batas bawah.
* Kategori yang memiliki bentuk distribusi memanjang merupakan kategori yang memilki varian besar. Sementara yang memiliki bentuk distribusi lebih pendek merupakan kategori yang memiliki varian kecil.
